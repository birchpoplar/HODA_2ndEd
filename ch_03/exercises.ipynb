# Exercise 1

```python
import pandas as pd

stocks = {}
ticks = ['aapl', 'amzn', 'fb', 'goog', 'nflx']

for tick in ticks:
    df = pd.read_csv('exercises/%s.csv' % tick)
    stocks[tick] = df
```

```python
for tick in ticks:
    stocks[tick] = stocks[tick].assign(ticker=tick.upper())
```

```python
df_faang = pd.concat([stocks[tick] for tick in ticks])
```

```python
df_faang.dtypes
```

```python
df_faang.to_csv('exercises/faang.csv',index=False)
```

# Exercise 2

```python
df_faang = df_faang.assign(
    date = pd.to_datetime(df_faang.date),
    volume = lambda x: x.volume.astype('int'))
```

```python
df_faang.dtypes
```

```python
df_faang = df_faang.sort_values(by=['date','volume']).reset_index()
```

# Exercise 6

```python
df_covid = pd.read_csv('exercises/covid19_cases.csv')
```

```python
df_covid.dtypes
```

```python
df_covid.head()
```

```python
df_covid = df_covid.assign(
    date = pd.to_datetime(df_covid.dateRep))
```

```python
df_covid.head()
```

```python
df_covid.set_index('date', inplace=True)
```

```python
df_covid.head()
```

```python
df_covid = df_covid.sort_index()
df_covid.head()
```

```python
df_covid_rpl = df_covid.replace({'United_States_of_America': 'USA', 'United_Kingdom': 'UK'})
df_covid_rpl.head()
```

```python
df_covid_rpl['countriesAndTerritories'].unique()
```

```python
countries = ['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA']
df_covid_rpl_filt = df_covid_rpl[df_covid_rpl['countriesAndTerritories'].isin(countries)]
df_covid_rpl_filt.head()
```

```python
pivoted_df = df_covid_rpl_filt.pivot(index='dateRep', columns='countriesAndTerritories', values='cases')
pivoted_df = pivoted_df.fillna(0)
pivoted_df.head()
```
